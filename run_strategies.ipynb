```python
# Name all of the price_periods
import os
path = r'.\\price_period_csv\\'
price_period_name_list = []
# Chose which price periods you want to look at
# price_periods = 'low'
price_periods = 'all'
# price_periods = '2022_price_data'

for _, _, files in os.walk(path):
    for file in files:
        if price_periods.lower() in file.lower() or price_periods.lower() == 'all':
            price_period_name_list.append(file)

print(f'price_periods found:\n{price_period_name_list}')

# Set default buy size to be 10k
starting_usd = 10000
# Seconds in a day
seconds_in_a_day = 60*60*24
```

```python
# Dollar Cost Average - Takes around 645 min
from specific_strategies import dca

# Number of days between dollar cost average buys
days_between_buys_list = [1/24, 1, 7, 14, 28]
# days_between_buys_list = [1]

# loop through time periods
for price_period_name in price_period_name_list:
    # loop through days_between_buys list
    for days in days_between_buys_list:
        # Turns days into seconds
        days = days*seconds_in_a_day
        dca_strategy = dca.base_dca(
            starting_usd=starting_usd,
            time_between_action=days,
            price_period_name=price_period_name
        )
        dca_strategy.run_logic()
```

```python
# All in start
from specific_strategies import all_in_start

# loop through time periods
for price_period_name in price_period_name_list:
        # we don't really need this but we just have to pick a number that won't be larger than our price_period
        days = 28
        # Turns days into seconds
        days = days*seconds_in_a_day
        all_in_strategy = all_in_start.base_all_in(
            starting_usd=starting_usd,
            time_between_action=days,
            price_period_name=price_period_name
        )
        all_in_strategy.run_logic()
```

```python
# All in top
from specific_strategies import all_in_top
# loop through time periods
for price_period_name in price_period_name_list:
        # The shorter the days, the better the accuracy for this.
        days = .1
        # Turns days into seconds
        days = days*seconds_in_a_day
        all_in_top_strategy = all_in_top.base_all_in_top(
            starting_usd=starting_usd,
            time_between_action=days,
            price_period_name=price_period_name
        )
        all_in_top_strategy.run_logic()
```

```python
# All in bottom
from specific_strategies import all_in_bottom
# loop through time periods
for price_period_name in price_period_name_list:
        # The shorter the days, the better the accuracy for this.
        days = .1
        # Turns days into seconds
        days = days*seconds_in_a_day
        all_in_bottom_strategy = all_in_bottom.base_all_in_bottom(
            starting_usd=starting_usd,
            time_between_action=days,
            price_period_name=price_period_name
        )
        all_in_bottom_strategy.run_logic()
```
